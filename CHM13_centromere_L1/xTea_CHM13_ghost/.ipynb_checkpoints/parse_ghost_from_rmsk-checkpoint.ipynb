```python
def parse_rmsk(sf_rmsk):
    m_rcds={}
    with open(sf_rmsk) as fin_rmsk:
        for line in fin_rmsk:
            fields=line.split()
            if len(fields)<1:
                continue
            if fields[0]=="score" or fields[0]=="SW":
                continue

            s_contig=fields[4]
            i_contig_start=int(fields[5])
            i_contig_end=int(fields[6])
            s_rc=fields[8]
            s_sub_family=fields[9]
            s_rep_family=fields[10]

            i_cns_start=-1
            i_cns_end=-1
            b_rc = False
            if s_rc=="+":
                i_cns_start=int(fields[11])
                i_cns_end=int(fields[12])
            else:
                i_cns_start=int(fields[13])
                i_cns_end=int(fields[12])
                b_rc=True

            hit_rcd=(i_contig_start, i_contig_end, b_rc, s_sub_family, s_rep_family, i_cns_start, i_cns_end)
            if s_contig not in m_rcds:
                m_rcds[s_contig]=[]
            m_rcds[s_contig].append(hit_rcd)
    return m_rcds

#file list is generate by command line:
#find . -maxdepth 2  -name "classified_results.txt.merged_LINE1.txt.fa.out" > tmp.list2
#grep -i "Alpha" $(cat tmp.list2) > LINE1_insetion_with_Alpha_sat_TD.txt
def slct_centromere_copies(m_rmsk_rcds):
    m_slct_alpha={}
    m_slct_beta={}
    m_slct_hsatii={}
    
    for s_contig in m_rmsk_rcds:
        for rcd in m_rmsk_rcds[s_contig]:
            fields=s_contig.split("_")
            s_ori=fields[-1]
            s_id="_".join(fields[:-1])
            s_sub_family=rcd[3]
            if "Alpha" in s_sub_family:#
#                         if s_id not in m_slct:
#                             m_slct[s_id]=
                m_slct_alpha[s_id]=1
            elif "Beta" in s_sub_family:
                m_slct_beta[s_id]=1
            elif "HSATII" in  s_sub_family:
                m_slct_hsatii[s_id]=1
    return m_slct_alpha, m_slct_beta, m_slct_hsatii
####
####            
sf_rmsk_list="/n/data1/hms/dbmi/park/simon_chu/projects/XTEA/xTEA_paper/long_reads/ghost_flank_rmsk/all_rslts.list"
m_sample_rslts={}
with open(sf_rmsk_list) as fin_list:
    for line in fin_list:
        fields=line.rstrip().split("/")
        s_fname=fields[-1]
        l_name_fields=s_fname.split("_")
        s_sample=l_name_fields[0]
        m_sample_rslts[s_sample]=line.rstrip()
####
####
with open("number_of_ghost_full_length_L1_by_samples.csv", "w") as fout_csv:
    fout_csv.write("Sample,Type,Count\n")
    for s_sample in m_sample_rslts:
        sf_rmsk=m_sample_rslts[s_sample]
        m_ctg=parse_rmsk(sf_rmsk)
        m_slct_alpha, m_slct_beta, m_slct_hsatii=slct_centromere_copies(m_ctg)
        n_alpha=len(m_slct_alpha)
        n_beta=len(m_slct_beta)
        n_hsatii=len(m_slct_hsatii)
        fout_csv.write(s_sample+",Alpha,"+str(n_alpha)+"\n")
        fout_csv.write(s_sample+",Beta,"+str(n_beta)+"\n")
        fout_csv.write(s_sample+",HSATII,"+str(n_hsatii)+"\n")
#
```
